```python
#IMPORTING LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# LOADING THE DATASET
data_diabetes = load_diabetes()
X = data_diabetes.data
Y = data_diabetes.target
```

```python
#TRAINING/TESTING
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#SCALING THE DATA
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#TRAINING THE MODEL
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#EVALUATION
Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
#PLOTTING (ACTUAL VS PREDICTION)
plt.scatter(Y_test, Y_pred, color="blue", edgecolor='k')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Diabetes Dataset - Actual vs Predicted")
plt.show()
```

```python
#EXAMPLE - NEW PATIENT PREDICTION 

input_data = (0.0380759, 0.0506801, 0.0616962, 0.0218724, -0.0442235, -0.0348208,
              -0.0434008, -0.00259226, 0.0199075, -0.0176461)  # Example patient data
input_data_as_array = np.asarray(input_data).reshape(1, -1)
input_data_scaled = scaler.transform(input_data_as_array)

prediction = model.predict(input_data_scaled)
if prediction[0] == 1:
    print("The patient has Diabetes.")
else:
    print("The patient does NOT have Diabetes.")
```
