```python
#COLLECTING THE DATA 

data='HousePricePrediction.xlsx'
```

```python
#READING THE DATA

df=pd.read_excel(data)
```

```python
#DATA PRE_PROCESSING(removal of):
#a-null values
#b-duplicate values
#c-unnecessary columns

df["SalePrice"]= df["SalePrice"].fillna(df["SalePrice"].mean())
```

```python
df
```

```python
#dropping remaining null values
df=df.dropna()
```

```python
df
```

```python
!pip install scikit-learn
```

```python
#IMPORTING LabelEncoder

from sklearn.preprocessing import LabelEncoder
```

```python
df
```

```python
#encoding object columns (object-integer)

obj_cols=['MSZoning','LotConfig','BldgType','Exterior1st']
le=LabelEncoder()
for col in obj_cols:
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
#features and target

X=df.drop(['SalePrice'],axis=1)
Y=df['SalePrice']
```

```python
df
```

```python
Y=df['SalePrice']
```

```python
#TRAINING AND SPLITTING 

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2, random_state=42, shuffle=True)
```

```python
df
```

```python
#SCALING DATA

from sklearn.preprocessing import StandardScaler
```

```python
scalar= StandardScaler()
X_train_scaled=scalar.fit_transform(X_train)
X_test_scaled=scalar.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, Y_train)
```

```python
#Prediction

Y_pred=model.predict(X_test_scaled)
df=Y_pred
```

df

```python
#EVALUATION

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(f"MAE : {mean_absolute_error(Y_test, Y_pred):,.2f}")
print(f"R^2 : {r2_score(Y_test, Y_pred):,.3f}")
```

```python
#SINGLE PREDICTION EXAMPLE
input_df =(0,60,3,8450,4,0,5,2003,2003,12,0,0)
input_df_array = np.asarray(input_df).reshape(1, -1)
input_df_scaled = scalar.transform(input_df_array)
predicted_price = model.predict(input_df_scaled)
print(f"Predicted House Price : {predicted_price[0]:,.2f}")
```
