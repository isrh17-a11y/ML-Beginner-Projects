```python
import random

def chatbot():
    print("Bot: Hi there! I'm 'MIMI ;)', your python chatbot. What can I help you with today? Type 'bye' to exit the conversation.")
    while True:
        user_input = input("You: ").lower()
        if user_input == "bye":
            print("Goodbye! Have a great day.")
            break
        elif "hello" in user_input or "hi" in user_input:
            print("Hello there! How can I help you today?")
        elif "how are you" in user_input:  
            print("I'm doing great, thanks for asking! I hope you're doing well too, what can I assist you with today?")
        elif "weather" in user_input or "what is the forecast today" in user_input or "is it going to rain today" in user_input:
            print("I'm afraid I can't help you with that but you can check the forecast for today using the weather app. Anything else you want me to help you with?")
        elif "fun" in user_input or "i wanna lighten my mood" in user_input:
            print("Sure! Something fun like a riddle, a tongue twister or rock paper scissors with a bot? ;) Anything you like!")
        elif "riddle" in user_input:
            print("Great! Here comes a riddle - I speak without a mouth and hear without ears. I have no body, but I come alive with the wind. What am I?")
        elif "echo" in user_input:
            print("Amazing! You guessed it right. Do you want to try something else?")
        elif "yes" in user_input:  
            print("How about rock-paper-scissors with a bot? I'm sure that would be better!")
        elif "sure" in user_input or "okay" in user_input or "yes" in user_input: 
            print("Alright! Let's play... in 1...2...3..")
        elif "scissors" in user_input or "rock" in user_input:
            print("ROCK!..oops. let's try again")
        elif "paper" in user_input:
            print("YOU WIN!")
        elif "tongue twister" in user_input:
            print("Sure! Try this - Pad kid poured curd pulled cod - Tough ain’t it ;)")
        else:  
            print("That's not quite right. Would you like to try again?")

chatbot()
```
